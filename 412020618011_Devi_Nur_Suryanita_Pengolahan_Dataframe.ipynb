```python
import pandas as pd
```

```python
#persiapan data frame dictionary

data = {'nama': ['Devi Nur', 'Denz sixteen', 'Danzly Sakurai'], 'tempat_kelahiran': ['Bogor, Jawa Barat', 'Ngawi, Jawa Timur', 'Osaka, Jepang']}
df = pd.DataFrame(data)
df
```

```python
#mau memecah kolom menjadi nama depan dan nama belakang
#di split berdasarkan spasi
#pararel expand=True untuk mengizinkan output berbentuk kolom / multiply column

df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

```python
#memecah nama kota dan provinsi
#yang membedakan dengan yang diatas adalah dengan memakai koma

df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

```python
#menggunakan unstack() pada multiple indexes dataframe
#menata ulang dataframe dengan multiple indexes menggunakan unstack()

import pandas as pd
```

```python
df = pd.read_csv('./titanicfull (1).csv')
df.head()
```

```python
#mengelompokkan data kedalam beberapa kelompok, disini kita kelompokkan berdasarkan sex dan pclass
#data yang akan kita amati yaitu data survived nya 1 untuk selamat 0 untuk tidak selamat
#rata2 nilai yang selamat
#to_frame untuk mempercantik tampilan

df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()
```

```python
#menata ulang dataframe dengan multiple indexes
#cara penempatan index sudah berbeda

df.groupby(['sex', 'pclass'])['survived'].mean().unstack()
```

```python
#resampling pada data deret waktu (time series)

import pandas as pd
import numpy as np
```

```python
#365 untuk hari, 24 untuk jam
#interval waktu berdasarkan jam

n_rows = 365 * 24
n_cols = 2
cols = ['kolom1', 'kolom2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq = 'H')
df
```

```python
#resampling data dengan interval monthly dengan .resample
#interval waktu berdasarkan bulan

df.resample('M')['kolom1'].sum().to_frame()
```

```python
#resampling data dengan interval dengan .daily
#interval waktu berdasarkan hari

df.resample('D')['kolom1'].sum().to_frame()
```
